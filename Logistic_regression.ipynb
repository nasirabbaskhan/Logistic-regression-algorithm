```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
```

```python
# import the data set from sklearn
from sklearn import datasets
```

```python
df = datasets.load_iris()
```

```python
df
```

```python
# creating the dataframe
df = pd.DataFrame({"sepal length":df.data[:,0],
               "sepal width":df.data[:,1],
               "petal length":df.data[:,2],
               "petal width":df.data[:,3],
               "species":df.target})
```

```python
df
```

```python
# for binary so remove 
df["species"] = df["species"].replace({0:"setosa", 1:"versicolor",2:"virginica"})
```

```python
df
```

```python
## for binary remove 'setosa

df = df[df["species"]!='setosa']
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["species"] = df["species"].map({"versicolor":0, "virginica":1})
```

```python
df
```

```python
x = df.iloc[:,:-1]
```

```python
x
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
# train test split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
# x_test
# y_train
# y_test
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
# playing with hyperparameters
from sklearn.model_selection import GridSearchCV
parameter = {'penalty':['l1','l2','elasticnet'], 'C':[1,2,3,4,5,6,8,10,20,30,40,50],"max_iter":[100,200,300]}
```

```python
classifier_regressor = GridSearchCV(classifier, param_grid=parameter,scoring='accuracy',cv=5)
```

```python
# train the model
classifier_regressor.fit(x_train,y_train)
```

```python
# which parameter combination is selected
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
# prediction
y_predict = classifier_regressor.predict(x_test)
y_predict
```

```python
y_test
```

```python
# accuracy score
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_predict, y_test)
print(score)
```

```python
print(classification_report(y_predict, y_test))
```

```python
sns.displot(y_test-y_predict, kind='kde')
```
